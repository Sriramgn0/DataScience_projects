# Money-Market Data Visualization

```python
# read in the data
df = pd.read_csv("../data/snb-data-zimoma-en-all-20170502_1430.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index(['Date', 'D0']).unstack()['Value']
term_order = ['SARON', '1TGT', '1M', 'EG3M', '3M0', '3M1', '3M2', '3M3', '3M4', '6M', '12M',]
df = df[term_order]
df.head()
```

## Plot labels and shading manually

```python
fig, ax = plt.subplots(figsize=(12, 8))
ser = df['1TGT']
ser.plot(ax=ax)

ax.xaxis.set_label_text("")
ax.yaxis.set_label_text("Bond Yield (%)")

start = '1973-11'
ypos = ser.loc[start]
ax.annotate('Oil Price Shock', 
             xy=(start, ypos),
             xytext=(start, ypos + 2),
             arrowprops=dict(arrowstyle="->", color='k'))

start = '1991-01'
end = '1993-12'
ax.axvspan(start, end, facecolor='k', alpha=0.2, label="CH Recession")

start = '2008-09'
text_start = '2002-09'
ypos = ser.loc[start]
ax.annotate('Lehman Brothers', 
             xy=(start, ypos),
             xytext=(text_start, ypos + 2),
             arrowprops=dict(arrowstyle="->", color='k'))

start = '2009-01'
end = '2009-12'
ax.axvspan(start, end, facecolor='k', alpha=0.2)

start = '2011-08'
end = '2015-01'
ax.axvspan(start, end, facecolor='g', alpha=0.2, label="CHF/EUR floor in place")
ax.legend(bbox_to_anchor=(1.02, 1), loc=2)
None
# grrr -- export to svg does not work on this file! Export to PDF and convert to svg in Inkscape
# plt.savefig("bond.pdf")
```

## Use helper functions to improve code readibility

```python
def add_text(df, ax, arrow_x, text_x, text_y_offset, text):
    arrow_y = df.loc[arrow_x]
    text_y = arrow_y + text_y_offset
    ax.annotate(text, 
                 xy=(arrow_x, arrow_y),
                 xytext=(text_x, text_y),
                 arrowprops=dict(arrowstyle="->", color='k'),
                 fontsize=14)
    
def add_span_text(ax, start, end, text_y, color, text=None):
    ax.axvspan(start, end, color=color, alpha=0.2, label=text)


fig, ax = plt.subplots(figsize=(12, 8))
ser = df['1TGT']
ser.plot(ax=ax)

ax.xaxis.set_label_text("")
ax.yaxis.set_label_text("Bond Yield (%)")

add_text(ser, ax, "1973-11", "1973-11", 2, 'Oil Price Shock')
add_span_text(ax, '1991-01', '1993-12', 15, 'k', 'Recession CH')
add_text(ser, ax, "2008-09", '2002-09', 2, 'Lehman Brothers')
add_span_text(ax, '2009-01', '2009-12', 15, 'k', None)
add_span_text(ax, '2011-08', '2015-01', 10, 'g', 'CHF/EUR floor in place')
ax.legend(bbox_to_anchor=(1.01, 1), loc=2)
plt.savefig("moneymarket.pdf")
None
```

# MPG Data Labeling

```python
df = pd.read_csv("../data/mpg.csv")
df['trans_type']= df['trans'].apply(lambda x: x.split('(')[0])
df.head()
```

**1. Plot displ vs hwy and label the cars with the best/worst hwy mileage**

```python
from matplotlib.colors import rgb_to_hsv, hsv_to_rgb
fig, ax = plt.subplots(figsize=(10, 7))
groups = df.groupby('class')
colors = sns.mpl_palette("Set1", 7)
color_class = {}
for (idx, (name, group)) in enumerate(groups):
    c_rgb = colors[idx]
    color_class[name] = c_rgb
    group.plot('displ', 'hwy', marker='.', linestyle='', c=c_rgb, ms=12, alpha=0.5, label=name, ax=ax)
    
max_hwy = df.iloc[df['hwy'].idxmax()]
ax.plot(max_hwy['displ'], max_hwy['hwy'], marker='.', ms=20, c=color_class[max_hwy['class']], alpha=1)
ax.annotate(max_hwy['model'], 
            xy=(max_hwy['displ'], max_hwy['hwy']), 
            xytext=(max_hwy['displ'] + 0.15, max_hwy['hwy']), 
            color=color_class[max_hwy['class']],
            fontsize=12)

min_hwy = df.iloc[df['hwy'].idxmin()]
ax.plot(min_hwy['displ'], min_hwy['hwy'], marker='.', ms=20, c=color_class[min_hwy['class']], alpha=1)
ax.annotate(min_hwy['model'], 
            xy=(min_hwy['displ'], min_hwy['hwy']), 
            xytext=(min_hwy['displ'] + 0.15, min_hwy['hwy']), 
            color=color_class[min_hwy['class']],
            fontsize=12)

ax.set_xlim([1, 8])
ax.set_ylim([10, 50])
_ = ax.legend()
```

**2. Facet by class, plot displ vs hwy, and label the cars with the best/worst hwy mileage in each class**

```python
fig, axs = plt.subplots(1, 7, figsize=(16, 4), sharex=True, sharey=True)
groups = df.groupby('class')

mean_displ = df.groupby('class').mean()['displ'].sort_values()
class_to_index = {}
for cls, idx in zip(mean_displ.index, range(len(mean_displ.index))):
    class_to_index[cls] = idx
for name, group in groups:
    index = class_to_index[name]
    ax = axs[index]
    ax.plot(group['displ'], group['hwy'], marker='.', linestyle='', ms=12, alpha=0.5, label=name)
    ax.set_xlim([1, 8])
    ax.set_ylim([10, 50])
    ax.set_title(name)
    if index < 1:
        ax.xaxis.set_label_text('Displacement')
        ax.yaxis.set_label_text('Highway Fuel Mileage')
    max_hwy = group.loc[group['hwy'].idxmax()]
    ax.plot(max_hwy['displ'], max_hwy['hwy'], marker='.', ms=20, alpha=1)
    ax.annotate("{}({})".format(max_hwy['model'], max_hwy['trans_type']),
                xy=(max_hwy['displ'], max_hwy['hwy']), 
                xytext=(max_hwy['displ'] + 0.2, max_hwy['hwy'] + 0.2))
    min_hwy = group.loc[group['hwy'].idxmin()]
    ax.plot(min_hwy['displ'], min_hwy['hwy'], marker='.', ms=20, alpha=1)
    ax.annotate("{}({})".format(min_hwy['model'], min_hwy['trans_type']), 
                xy=(min_hwy['displ'], min_hwy['hwy']), 
                xytext=(min_hwy['displ'] - 0.2, min_hwy['hwy'] - 0.2))
```

```python
# Same as above, done in Seaborn
palette = sns.color_palette()
def facet_plot(x, y, **kwargs):
    ax = plt.gca()
    df = kwargs.pop("data")
    df.plot.scatter(x=x, y=y, ax=ax, alpha=0.5, **kwargs)
    max_hwy = df.loc[df['hwy'].idxmax()]
    trans = max_hwy['trans_type']
    ax.plot(max_hwy['displ'], max_hwy['hwy'], marker='.', c=palette[1], ms=20, alpha=1)
    ax.annotate("{}({})".format(max_hwy['model'], trans),
                xy=(max_hwy['displ'], max_hwy['hwy']), 
                xytext=(max_hwy['displ'] + 0.2, max_hwy['hwy'] + 0.2))
    min_hwy = df.loc[df['hwy'].idxmin()]
    ax.plot(min_hwy['displ'], min_hwy['hwy'], marker='.', c=palette[2], ms=20, alpha=1)
    ax.annotate("{}({})".format(min_hwy['model'], min_hwy['trans_type']), 
                xy=(min_hwy['displ'], min_hwy['hwy']), 
                xytext=(min_hwy['displ'] - 0.2, min_hwy['hwy'] - 0.2))    
g = sns.FacetGrid(df, col='class', col_wrap=4)
_ = g.map_dataframe(facet_plot, 'displ', 'hwy')
```

# Anscombe's Quartet

```python
df = pd.read_csv('../data/anscombe.csv')
df.head()
```

```python
df.groupby('dataset').mean()
```

```python
import statsmodels.formula.api as smf

def fit_and_predict(df):
    lm = smf.ols(formula="y ~ x", data=df).fit()
    x_bounds = [df['x'].min(), df['x'].max()]
    preds_input = pd.DataFrame({'x': x_bounds})
    predictions = lm.predict(preds_input)
    return lm, pd.DataFrame({'x': x_bounds, 'y': predictions})

fig, axs = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(8, 6))
for index, (name, tdf) in enumerate(df.groupby('dataset')):
    col = index % 2
    row = (index - col) // 2
    ax = axs[row][col]
    ax.scatter(tdf['x'], tdf['y'])
    lm, pred = fit_and_predict(tdf)
    ax.plot(pred['x'], pred['y'])
    ax.text(14, 10.5, "$r^2$ = {:.2f}".format(lm.rsquared))
    ax.text(11, 5, "slope = {:.1f}".format(lm.params.x))
    ax.text(11, 4, "intcpt = ({:.0f},0)".format(lm.params.Intercept))
    ax.set_title(name)
```
