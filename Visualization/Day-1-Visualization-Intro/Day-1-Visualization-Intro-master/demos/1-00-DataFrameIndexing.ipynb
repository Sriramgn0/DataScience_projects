# Data Frame Indexing

Indxing is not not explicitly part of the module, but potentially helpful for doing the exercises.

```python
import pandas as pd
import numpy as np

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Read in the frame
df = pd.read_csv("../data/snb-data-rendoblim-en-all-20170502_1430.csv")
df['Date'] = pd.to_datetime(df['Date'])
df.set_index(['Date', 'D0'], inplace=True)
df.sort_index(inplace=True)
df.head()
```

```python
# Stack and reorder the colums
cols = ["{}J".format(idx) for idx in range(1, 11)]
cols.extend(['20J', '30J'])
df = df.unstack()['Value'][cols]
df.head()
```

```python
# Take just data from 1990
df.loc["1990"].head()
```

```python
# Take just data from 1990 to 1992
df.loc["1990":"1992"].tail()
```

```python
# Take just data 3J - 5J data
df.loc[:,'3J':'5J'].head()
```

```python
# Same as above, using a slice
df.loc[:,slice('3J','5J')].head()
```

```python
# Take just data from 1990 to 1992 and 3J to 5J series
df.loc["1990":"1992", "3J":"5J"].tail()
```

```python
# Just January data
df.loc[lambda x: x.index.month == 1, :]
```
