```python
# Making the imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)

# Preprocessing Input data
data = pd.read_csv('Salary_Data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
plt.scatter(X, Y)
plt.show()
```

```python
# Building the model
B = 0
B_0 = 0

L = 0.01  # The learning Rate
epochs = 10  # The number of iterations to perform gradient descent

n = float(len(X)) # Number of elements in X

# Performing Gradient Descent 
for i in range(epochs): 
    Y_pred = B*X + B_0  # The current predicted value of Y
    D_B = (-2/n) * sum(X * (Y - Y_pred))  # Derivative wrt B
    D_B0 = (-2/n) * sum(Y - Y_pred)  # Derivative wrt B0
    B = B - L * D_B  # Update B
    B_0 = B_0 - L * D_B0  # Update B0
    plt.scatter(X, Y)
    plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
    plt.show()
print (B, B_0)
```

```python
# Making predictions
Y_pred = B*X + B_0

plt.scatter(X, Y)
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
```
