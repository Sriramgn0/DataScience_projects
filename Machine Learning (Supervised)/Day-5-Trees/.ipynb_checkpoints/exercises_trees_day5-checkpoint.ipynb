```python
import sklearn
import numpy as np
import pandas as pd
from sklearn.utils import shuffle
from typing import Dict, Tuple, Callable
from sklearn import svm
from sklearn.model_selection import cross_val_score
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
cat_columns = ["diagnosis"]
#datap is processed data
datap = pd.get_dummies(data, prefix_sep="_",
                              columns=cat_columns)
```

```python
datap.head()
```

```python
datap = datap.drop(columns=['diagnosis_B'])
```

```python
corr = datap.corr()
corr.style.background_gradient(cmap='coolwarm')
```

Separate the data into the features (X) and targets (y). The target, also known as the label, is the value we want to predict, in this case the malignant diagnosis and the features are all the columns the model uses to make a prediction. We will also convert the Pandas dataframes to Numpy arrays because that is the way the algorithm works. (I save the column headers, which are the names of the features, to a list to use for later visualization).

```python
X = datap.loc[:, datap.columns != 'diagnosis_M']
y = datap['diagnosis_M'].values
```

```python
# create training and testing data sets which we will use later after we select a model from k-1 cross validation
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

Before we can make and evaluate predictions, we need to establish a baseline, a sensible measure that we hope to beat with our model. If our model cannot improve upon the baseline, then it will be a failure and we should try a different model or admit that machine learning is not right for our problem. The baseline prediction for our case can be the historical max temperature averages. In other words, our baseline is the error we would get if we simply predicted the average diagnosis.

```python
# Import the model we are using
from sklearn.ensemble import RandomForestClassifier
# Instantiate model with 1000 decision trees
rf = RandomForestClassifier(n_estimators = 400)
# Train the model on training data
rf.fit(X_train, y_train);
```

```python
# Use the forest's predict method on the test data
predictions = rf.predict(X_test)
#check accuracy of our model on the test data and confirm it's working
rf.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
#create new a knn model
rf = RandomForestClassifier()
#create a dictionary of all values we want to test for n_neighbors
param_grid = {"n_estimators": [10,100,250,500,1000]}
#use gridsearch to test all values for degrees
rf_gs = GridSearchCV(rf, param_grid, cv=5, scoring='roc_auc',)
#fit model to data
rf_gs.fit(X_train, y_train)
```

```python
rf_gs.best_score_
```

```python
rf_gs.best_estimator_.fit(X_train, y_train)
```

```python
rf_gs.best_estimator_
```

```python
# Import tools needed for visualization
from sklearn.tree import export_graphviz
import pydot
# Pull out one tree from the forest
tree = rf_gs.best_estimator_.estimators_[5]
# Export the image to a dot file
export_graphviz(tree, out_file = 'tree.dot', feature_names = X_train.columns, rounded = True, precision = 1)
# Use dot file to create a graph
(graph, ) = pydot.graph_from_dot_file('tree.dot')
# Write graph to a png file
graph.write_png('tree_classifiers.png')
```

```python
from sklearn.model_selection import RandomizedSearchCV
#create new a knn model
rf = RandomForestClassifier()
#create a dictionary of all values we want to test for n_neighbors
param_grid = {
    "n_estimators": [10,100,250,500,1000],
    "max_depth":[4,5,6,7,8],
    "min_samples_split":[2,3,4,5],
    "min_samples_leaf":[1,2,3],
    "criterion":['gini','entropy']}
#use gridsearch to test all values for degrees
rf_rgs = RandomizedSearchCV(rf, param_grid, cv=5, scoring='roc_auc',n_iter=500,verbose=2, n_jobs=-1)
#fit model to data
rf_rgs.fit(X_train, y_train)
```

```python
rf_rgs.best_score_
```

```python
rf_rgs.best_estimator_
```

```python
best_model = rf_rgs.best_estimator_

best_model.fit(X_train, y_train)

y_pred = best_model.predict_proba(X_test)
```

```python
plt.hist(y_pred)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
rforrest_roc_auc = roc_auc_score(y_test, y_pred[:,1])
fpr, tpr, thresholds = roc_curve(y_test, y_pred[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % rforrest_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

Create plots to understand the data further:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.feature_selection import SelectKBest
```

```python
rf = RandomForestClassifier()
#pipeline to include feature selection and model selection
pipeline = Pipeline([('kbest',SelectKBest()),('classifier', rf)])


param_grid = {
    "classifier__n_estimators": [10,100,250,500,1000],
    "classifier__max_depth":[4,5,6,7,8],
    "classifier__min_samples_split":[2,3,4,5],
    "classifier__min_samples_leaf":[1,2,3],
    "classifier__criterion":['gini','entropy'],
    "kbest__k":[5,10,30]}


rf_rg_pipe = RandomizedSearchCV(pipeline, param_grid, cv=5, scoring='roc_auc',n_iter=50,verbose=2, n_jobs=-1)

rf_rg_pipe.fit(X_train,y_train)





```

```python
rf_rg_pipe.best_score_
```

```python
rf_rg_pipe.best_estimator_
```

```python
import pandas as pd
%matplotlib inline
feats = {} # a dict to hold feature_name: feature_importance
for feature, importance in zip(data.columns, rf_rg_pipe.best_estimator_.named_steps['classifier'].feature_importances_):
    feats[feature] = importance #add the name/value pair 
importances = pd.DataFrame.from_dict(feats, orient='index').rename(columns={0: 'Gini-importance'})
importances.sort_values(by='Gini-importance',ascending = False).plot(kind='bar', rot=45,figsize=(20,10))
```
