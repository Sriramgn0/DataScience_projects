```python
import pandas as pd
```

```python
reviews = pd.read_csv("ign.csv")
```

```python
reviews.head() #when no input given, default values is 5
#alternatively we could use reviews.iloc[:5] which would give the first 5 results as well
```

```python
reviews.shape  #shape of our data in format --> (rows, columns)
```

```python
reviews.iloc[0:5,:] 
```

```python
#notice the first column 'Unnamed' isn't helpful so we can remote that column
reviews = reviews.iloc[:,1:]
reviews.head()
```

```python
"""
A major advantage of Pandas over NumPy is that each of the columns and rows has a label. 
Working with column positions is possible, but it can be hard to keep track of which number corresponds to 
which column.

""" 
reviews.loc[0:5,:] #looking at the first 5 rows
```

```python
reviews.index 
```

```python
some_reviews = reviews.iloc[10:20,] #let's grab just a few reviews. Let's set some_reviews to only have entries 10-19
```

```python
some_reviews.head() #this will grab the first 5 entries by default from some_reviews (rows 10-14)
```

```python
#instad of specifiying column numbers, we can call them by column names directly.
reviews.loc[:5,['score','release_day']]   #note, the output is only "pretty" if you call more than 1 column.
```

```python
# the easiest way though to call a column is simply to put the column name in brackets
reviews["score"]
```

```python
#likewise we can specifiy multiple columns
reviews[["score", "release_year"]]
```

```python
"""
When we retrieve a single column, we’re actually retrieving a Pandas Series object. 
A DataFrame stores tabular data, but a Series stores a single column or row of data.
"""
type(reviews["score"])  #let's proove to ourselves the columns are series type
```

```python
s1 = pd.Series([1,2]) #let's create new columns to see how series work
```

```python
#find the mean of specific column
reviews["score"].mean()
```

```python
#find the mean of all columns with numerical data
reviews.mean() #By default, axis is equal to 0, and will compute the mean of each column
#reviews.mean() == reviews.mean(0)
```

```python
#find the mean values for all rows with numerical data
reviews.mean(axis=1)
```
