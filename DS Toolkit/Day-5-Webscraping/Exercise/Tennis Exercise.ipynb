# 1. How long has Roger Federer been in the top 5?

```python
url = 'https://www.atptour.com/en/players/roger-federer/f324/rankings-history'
```

```python
# get the content from the link
response = requests.get(url, timeout=50)

# parse the content with BeautifulSoup with the html.parser method
soup = BeautifulSoup(response.content, "html.parser")

#check status code of the downloaded page
response.status_code
```

```python
soup
```

```python
# we find the right path to get the link for every job
soup.find('tbody')
```

```python
i=5
date = soup.find('tbody').findAll('tr')[i].findAll('td')[0].get_text().strip()
singles_rank = soup.find('tbody').findAll('tr')[i].findAll('td')[1].get_text().strip()
date,singles_rank
```

```python
len(soup.find('tbody').findAll('tr'))
```

```python
# initiating a dataframe to store the information
columns = [
    'Date',
    'Ranking',
]
df = pd.DataFrame(columns = columns)
```

```python
for i in range(len(soup.find('tbody').findAll('tr'))):
    date = soup.find('tbody').findAll('tr')[i].findAll('td')[0].get_text().strip()
    singles_rank = soup.find('tbody').findAll('tr')[i].findAll('td')[1].get_text().strip()


    
    # we append the data to the dataframe
    df = df.append({
    'Date':date,
    'Ranking':singles_rank
    }, ignore_index=True)
    # before the end of the loop we introduce a time delay to respect others that want to access the website
```

```python
#df.replace('T', " ")
df['Ranking'] = df['Ranking'].map(lambda x: x.lstrip('T').rstrip('T'))
```

```python
df['Ranking'] = df['Ranking'].astype(int)
```

```python
df = df.drop(df[df.Ranking > 5].index)
```

```python
df = df.drop(df[df.Ranking == 0].index)
```
