# 1. Which player hit the most home runs in 2002?

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT
    playerid,
    hr
FROM
    batting
WHERE yearid = 2002
ORDER BY
    hr DESC
LIMIT 1;
    """, con)

df_sql
```

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT batting.playerid, master.namefirst, master.namelast, batting.yearid, batting.hr
FROM master
INNER JOIN batting on batting.playerid = master.playerid
WHERE hr = 57 and yearid = 2002

    """, con)

df_sql
```

# 2. Which team spent the most/least money on player salaries in 2002?

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT
    playerid,
    salary,
    teamid,
    yearid
FROM
    salaries
WHERE yearid = 2002
ORDER BY
    salary DESC
LIMIT 10
    """, con)

df_sql
```

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT salaries.teamid, 
SUM (salary) 
FROM salaries
WHERE yearid = 2002
GROUP BY teamid
ORDER BY SUM DESC
LIMIT 1;
    """, con)
df_sql
```

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT salaries.teamid,
SUM (salary) 
FROM salaries
WHERE yearid = 2002
GROUP BY teamid
ORDER BY SUM
LIMIT 1;
    """, con)
df_sql
```

# 3. Which player averaged the fewest at bats between home runs in 2002?

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT batting.playerid, master.namefirst, master.namelast, batting.yearid, batting.hr, batting.ab,ab/hr AS AB_HR_Ratio
FROM master
INNER JOIN batting on batting.playerid = master.playerid
WHERE yearid = 2002 and hr > 1
--GROUP BY playerid
ORDER BY AB_HR_Ratio
limit 1
    """, con)

df_sql
```

# 4. Which player in 2002 had the highest on-base percentage?

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT batting.playerid, master.namefirst, master.namelast, batting.yearid, batting.h, batting.ab,batting.bb,batting.hbp,batting.sf,((H+BB+HBP)/(AB+BB+HBP+SF)) AS base_perc
FROM master
INNER JOIN batting on batting.playerid = master.playerid
WHERE yearid = 2002 and (AB+BB+HBP+SF) !=0
ORDER BY base_perc DESC
LIMIT 10
    """, con)

df_sql
```

# 5. Which Yankees pitcher had the most wins in a season in the 00’s?

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT pitching.teamid, pitching.playerid, pitching.w, pitching.yearid, master.namefirst, master.namelast
from master
INNER JOIN pitching on pitching.playerid = master.playerid
WHERE yearid >= 2000 and teamid = 'NYA'
ORDER by pitching.w DESC
LIMIT 10
""", con)

df_sql
```

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT pitching.teamid, pitching.playerid, pitching.w, pitching.yearid, master.namefirst, master.namelast, fielding.pos
from master
INNER JOIN pitching
ON master.playerid = pitching.playerid
INNER JOIN fielding
ON master.playerid = fielding.playerid
WHERE pitching.yearid >= 2000 and pitching.teamid = 'NYA'and pos = 'P'
ORDER by pitching.w DESC
LIMIT 10
""", con)

df_sql
```

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT teams.name, pitching.teamid, pitching.playerid, pitching.w, pitching.yearid, master.namefirst, master.namelast, fielding.pos
from master
INNER JOIN pitching
ON master.playerid = pitching.playerid
INNER JOIN fielding
ON master.playerid = fielding.playerid
INNER JOIN teams
ON pitching.teamid = teams.teamid
WHERE pitching.yearid >= 2000 and teams.name = 'New York Yankees' and pos = 'P'
ORDER by pitching.w DESC

""", con)

df_sql
```

# 7. In the 2000’s, did the Yankees draw more or fewer walks (Base-on-Balls or BB) as the decade went on?

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT batting.teamid, batting.yearid, batting.bb, teams.name
FROM batting
INNER JOIN teams
ON batting.teamid = teams.teamid
WHERE batting.yearid >= 2000 and teams.name = 'New York Yankees'
ORDER by batting.bb DESC
LIMIT 10
    """, con)

df_sql
```

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT batting.yearid, AVG(batting.bb) AS avg_bb
FROM batting
INNER JOIN teams
ON batting.teamid = teams.teamid
WHERE batting.yearid >= 2000 and teams.name = 'New York Yankees'
GROUP BY batting.yearid, batting.bb
ORDER by batting.bb DESC
LIMIT 10
    """, con)

df_sql
```

```python
df_sql = pd.read_sql_query("""
SET search_path = lahman2017;
SELECT batting.teamid, batting.yearid, AVG(batting.bb) AS avg_bb, teams.name
FROM batting
INNER JOIN teams
ON batting.teamid = teams.teamid
WHERE batting.yearid >= 2000 and teams.name = 'New York Yankees'
GROUP BY batting.yearid, batting.teamid,batting.bb,teams.name
ORDER by batting.yearid DESC
    """, con)

df_sql

```
